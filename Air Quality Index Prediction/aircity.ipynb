```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from datetime import datetime

data = pd.read_csv("city_day.csv")
```

```python
data=data.drop(columns=["AQI_Bucket"])
```

```python
data.head()
```

```python
data=data.dropna()
```

```python
data.info()
```

```python

cities = data['City'].unique()


for city in cities:
    data[f'City_{city}'] = (data['City'] == city).astype(int)


data['Date'] = pd.to_datetime(data['Date'],format="%d-%m-%Y")

data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
features = ['Year', 'Month', 'Day', 'PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene'] + [col for col in data.columns if 'City_' in col]
X = data[features]
y = data['AQI']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = RandomForestRegressor(n_estimators=300, max_depth=30, random_state=0)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("R2 Score:", r2)
```

```python

city_name = 'Hyderabad'
city_data = pd.DataFrame({
    'Year': [2024],  
    'Month': [4],    
    'Day': [18],    
    'PM2.5': [20],   
    'PM10': [30],    
    'NO': [10],     
    'NO2': [15],     
    'NOx': [18],    
    'NH3': [5],      
    'CO': [0.5],     
    'SO2': [8],      
    'O3': [25],      
    'Benzene': [1],  
    'Toluene': [1.5],
    'Xylene': [2],  
})
for city in cities:
    if city == city_name:
        city_data[f'City_{city}'] = 1
    else:
        city_data[f'City_{city}'] = 0

predicted_aqi = model.predict(city_data)

print("Predicted AQI for", city_name, ":", predicted_aqi[0])
```
